```python
import nnunetv2, os
# Ghi trainer vào đúng package nnU-Net để lệnh `-tr` tìm thấy (không cần restart kernel vì train chạy ở subprocess)
trainer_dir = os.path.join(os.path.dirname(nnunetv2.__file__), 'training', 'nnUNetTrainer')

trainer_code = r'''
import os, torch
from torch.optim import AdamW
from nnunetv2.training.lr_scheduler.polylr import PolyLRScheduler
from nnunetv2.training.nnUNetTrainer.nnUNetTrainer import nnUNetTrainer

class nnUNetTrainer_HPO(nnUNetTrainer):
    """Mọi siêu tham số đọc từ biến môi trường -> Optuna chỉ cần set env rồi gọi train."""
    def __init__(self, plans, configuration, fold, dataset_json, device=torch.device('cuda')):
        super().__init__(plans, configuration, fold, dataset_json, device)
        self.initial_lr                  = float(os.environ.get('HPO_LR', '1e-2'))
        self.weight_decay                = float(os.environ.get('HPO_WD', '3e-5'))
        self.oversample_foreground_percent = float(os.environ.get('HPO_OVERSAMPLE', '0.33'))
        self.num_epochs                  = int(os.environ.get('HPO_EPOCHS', '50'))    # proxy ngắn
        self.num_iterations_per_epoch    = int(os.environ.get('HPO_ITERS', '125'))   # mặc định 250
        self.num_val_iterations_per_epoch= int(os.environ.get('HPO_VAL_ITERS', '25'))
        self._hpo_opt = os.environ.get('HPO_OPT', 'SGD')

    def configure_optimizers(self):
        if self._hpo_opt.lower() == 'adamw':
            optimizer = AdamW(self.network.parameters(), lr=self.initial_lr,
                              weight_decay=self.weight_decay, amsgrad=True)
        else:  # SGD mặc định của nnU-Net
            optimizer = torch.optim.SGD(self.network.parameters(), self.initial_lr,
                                        weight_decay=self.weight_decay, momentum=0.99, nesterov=True)
        return optimizer, PolyLRScheduler(optimizer, self.initial_lr, self.num_epochs)
'''
open(os.path.join(trainer_dir, 'nnUNetTrainer_HPO.py'), 'w').write(trainer_code)
print("Đã ghi trainer vào:", trainer_dir)
```

```python
import subprocess

SHOW = ("Epoch ", "train_loss", "val_loss", "Pseudo dice", "New best")  # dòng muốn hiện live

def _run_stream(cmd, env, logpath, prefix=""):
    """Lưu FULL log vào file, đồng thời in các dòng tiến độ ra notebook ngay khi train chạy."""
    env = dict(env); env["PYTHONUNBUFFERED"] = "1"     # ⭐ ép nnU-Net xả stdout theo từng dòng
    with open(logpath, "w") as lf:
        p = subprocess.Popen(cmd, env=env, stdout=subprocess.PIPE,
                             stderr=subprocess.STDOUT, text=True, bufsize=1)
        for line in p.stdout:
            lf.write(line); lf.flush()
            if any(k in line for k in SHOW):
                print(prefix + line.rstrip(), flush=True)
        p.wait()
    return p.returncode

def objective(trial):
    params = {
        "lr":           trial.suggest_float("lr", 1e-4, 3e-2, log=True),
        "weight_decay": trial.suggest_float("weight_decay", 1e-6, 1e-3, log=True),
        "oversample":   trial.suggest_float("oversample", 0.33, 0.80),
        "optimizer":    trial.suggest_categorical("optimizer", ["SGD", "AdamW"]),
    }
    print(f"\n{'='*64}\n▶ TRIAL {trial.number} BẮT ĐẦU | {params}\n{'='*64}", flush=True)

    env = dict(os.environ)
    env.update({"HPO_LR": str(params["lr"]), "HPO_WD": str(params["weight_decay"]),
                "HPO_OVERSAMPLE": str(params["oversample"]), "HPO_OPT": params["optimizer"],
                "HPO_EPOCHS": "50", "HPO_ITERS": "125", "HPO_VAL_ITERS": "25"})
    shutil.rmtree(out_folder, ignore_errors=True)

    log = f"/content/hpo_trial_{trial.number}.log"
    _run_stream(["nnUNetv2_train", DATASET, CONFIG, str(FOLD), "-tr", TRAINER],
                env, log, prefix=f"   [T{trial.number}] ")

    summary = os.path.join(out_folder, f"fold_{FOLD}", "validation", "summary.json")
    if not os.path.isfile(summary):
        print(f"❌ [Trial {trial.number}] FAIL — xem {log}"); print(open(log).read()[-2000:])
        raise optuna.TrialPruned()
    dice = json.load(open(summary))["foreground_mean"]["Dice"]
    print(f"✓ [Trial {trial.number}] XONG | Dice={dice:.4f}", flush=True)
    return dice
```

```python
import optuna, os, shutil, subprocess

# ==== phải KHỚP với cell HPO ====
DATASET = "Dataset001_KneeOA"
CONFIG  = "3d_fullres"
PLANS   = "nnUNetPlans"
TRAINER = "nnUNetTrainer_HPO"
STUDY   = "nnunet_kneeoa_hpo"
STORAGE = "sqlite:////content/drive/MyDrive/optuna_nnunet.db"  # đúng path đã dùng ở cell HPO
# =================================

# set lại env (mất sau restart)
os.environ['nnUNet_raw']          = '/content/nnUNet_raw'
os.environ['nnUNet_preprocessed'] = '/content/nnUNet_preprocessed'
os.environ['nnUNet_results']      = '/content/nnUNet_results'

# >>> FIX: nạp study từ DB, KHÔNG dùng biến `study` trong RAM <<<
study = optuna.load_study(study_name=STUDY, storage=STORAGE)

done = [t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE]
assert done, "Chưa có trial nào COMPLETE -> chạy cell HPO trước đã."
bp = study.best_params
print(f"✓ Best Dice (proxy)={study.best_value:.4f} | params={bp}")

# Xoá kết quả HPO ở fold_0, nếu không nnU-Net thấy checkpoint_final.pth sẽ tưởng đã xong và BỎ QUA
out_folder = os.path.join(os.environ['nnUNet_results'], DATASET, f"{TRAINER}__{PLANS}__{CONFIG}")
shutil.rmtree(out_folder, ignore_errors=True)

# Train ~300 epoch trên cả 5 fold với param tốt nhất
env = dict(os.environ)
env.update({"HPO_LR": str(bp["lr"]), "HPO_WD": str(bp["weight_decay"]),
            "HPO_OVERSAMPLE": str(bp["oversample"]), "HPO_OPT": bp["optimizer"],
            "HPO_EPOCHS": "300", "HPO_ITERS": "250", "HPO_VAL_ITERS": "50"})
for fold in range(5):
    print(f"\n===== Training fold {fold} (300 epoch) =====")
    subprocess.run(["nnUNetv2_train", DATASET, CONFIG, str(fold), "-tr", TRAINER], env=env)
```
